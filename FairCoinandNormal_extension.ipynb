# Tossing a fair coin

A simple experiment to show that tossing a **fair coin** several times will give a normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Below a simple function to simulate the tossing of a coin:

```python
def toss(n):
    r=np.random.rand(1,n)
    #r can be only a number between 0 and 1. If the value is less than 0.5 is 0=head
    # otherwise is 1 =tail (arbitrary, you can change that) 
    return np.round(r) #round to 1 sig.fig
```

Below is a function describing the experiment that Magda provided. ie tossing a coin many times and 

```python

def experiment(n_exp, N):
  results=[]
  for i in range(n_exp):
      r=toss(N)
      results.append(np.sum(r)/N)
  return results
```

Lets see how the distribution changes when we run the experiment several times with a small number of tosses, say 100.

```python
n_experiments = 100
n_tosses = 500

fig = plt.figure(figsize=(10,10))

results1 = experiment(n_experiments, n_tosses)
ax1 = fig.add_subplot(311)
ax1.hist(results1)
results2 = experiment(n_experiments, n_tosses)
ax2 = fig.add_subplot(312)
ax2.hist(results2)
results3 = experiment(n_experiments, n_tosses)
ax3 = fig.add_subplot(313)
ax3.hist(results3)

fig.suptitle('Distribution with '+str(n_tosses)+' tosses and '+str(n_experiments)+' experiements.',fontsize=14)
plt.xlabel('Number of tails')
ax1.set_ylabel('Frequency')
ax2.set_ylabel('Frequency')
ax3.set_ylabel('Frequency')
plt.show()

print("median=",np.median(results1+results2+results3))    
```

See how the distribution changes with a small number of experiments.  
Now lets see what happens when we run the expeiment 100,000 times.

```python
n_experiments = 100000
n_tosses = 500

fig = plt.figure(figsize=(10,10))

results1 = experiment(n_experiments, n_tosses)
ax1 = fig.add_subplot(311)
ax1.hist(results1)
results2 = experiment(n_experiments, n_tosses)
ax2 = fig.add_subplot(312)
ax2.hist(results2)
results3 = experiment(n_experiments, n_tosses)
ax3 = fig.add_subplot(313)
ax3.hist(results3)

fig.suptitle('Distribution with '+str(n_tosses)+' tosses and '+str(n_experiments)+' experiements.',fontsize=14)
plt.xlabel('Number of tails')
ax1.set_ylabel('Frequency')
ax2.set_ylabel('Frequency')
ax3.set_ylabel('Frequency')
plt.show()

print("median=",np.median(results1+results2+results3))    
```

Now we see that when we run the experiement a significant number of time, 100,000, we always get a normal distribution. This is why Magda was suggesting that in order to guarantee a normal distribution with coin tosses, we need to run the experiment over and over.
